```python
%matplotlib inline
import pickle
import numpy
from yellowbrick.text import TSNEVisualizer
from sklearn.feature_extraction.text import TfidfVectorizer
numpy.random.seed(42)
```

```python
words_file = "my_dataset.pkl"
labels_file = "my_feature_list.pkl"
word_data = pickle.load( open(words_file, "rb"))
labels_data = pickle.load( open(labels_file, "rb") )
```

```python
from sklearn import model_selection
features_train, features_test, labels_train, labels_test = model_selection.train_test_split\
    (word_data, labels_data, test_size=0.9, random_state=42)
```

```python
tfidf  = TfidfVectorizer(sublinear_tf=True, max_df=0.5, stop_words='english')
docs   = tfidf.fit_transform(features_train)
labels = labels_train
```

```python

# Create the visualizer and draw the vectors 
tsne = TSNEVisualizer()
tsne.fit(docs, labels)
tsne.poof()
```

## Frequency Distribution Visualization

```python
from yellowbrick.text.freqdist import FreqDistVisualizer 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')

vectorizer = CountVectorizer(stop_words = 'english')
docs       = vectorizer.fit_transform(features_train)
features   = vectorizer.get_feature_names()

visualizer = FreqDistVisualizer(features=features)
visualizer.fit(docs)
visualizer.poof()
```
